```python
import pandas as pd
import pickle
data  = pd.read_csv('/content/drive/MyDrive/Recipe/TB_RECIPE_SEARCH-210630.csv', encoding='cp949')
```

```python
data.isnull().sum()
```

```python
data1 = data.dropna(subset=['CKG_MTRL_CN'])
data1=data1.reset_index()
```

#1차 전처리

```python
import re
pattern1 = r'\[[^]]*\]'
pattern2 = r'\([^)]*\)'
pattern3 = r'\{[^)]*\}'
lst = []
lst2 = []
for i in data1.index:
   
  text = re.sub(pattern=pattern1, repl='|', string= data1.loc[i]['CKG_MTRL_CN'])
  text = re.sub(pattern=pattern2, repl='', string= text)
  text = re.sub(pattern=pattern3, repl='', string= text)
  text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘\(\)\[\]\<\>`\'…》]','', text)
  text = re.sub(pattern=r'\d.', repl='', string= text)
  text = re.sub(pattern=r'[a-z]', repl = '', string = text)
  for j in text.split('|'):
      co = list(filter(None, j.split(' ')))
      if co == []:
         continue
      lst.append(co[0])
      a = ','.join(lst)
  lst2.append(a)
  lst = []  

```

#2차 전처리

```python
#식재료명 사전 불러오기
with open('/content/drive/MyDrive/Recipe/food_list.p', 'rb') as f:
   food = pickle.load(f)
```

```python
lst3 = []
for j in range(len(lst2)):
  lst = []
  for i in lst2[j].split(','):
    if i in food:
      lst.append(i)
  lst3.append(lst)
```
