```python
import pandas as pd
```

```python
df = pd.read_csv('recipe_haemuk.csv', index_col=0)
# df.head()
```

```python
df = df.drop('recipe step', axis=1)
# df.head()
```

```python
# name/ ingredients/ ingredients_pre/ url
df = df.rename(columns = {'title': 'name'})
```

```python
# 몇 인분인지 앞의 숫자만 남기기

ls = []
for n in df['serving']:
    numbers = n[:2]
    if '인' in numbers:
        numbers = n[:1]
    else:
        pass
    ls.append(numbers)
```

```python
df['serve'] = ls
```

```python
df = df.drop(['serving'], axis=1)
# df.head()
```

```python
# df[serve]에서 '~, -'와 같은 특수 문자 제거
df['serve'] = df['serve'].str.extract(r'(\d+)')
```

```python
# 재료 맨 앞/뒤에 있던 '|' 불필요한 문자 제거
for i in range(len(df['ingredients'])):
    df['ingredients'][i] = df['ingredients'][i][1:-1]
```

```python
# 의미 없는 데이터 삭제
promo = df[df['serve'] == '0'].index
df = df.drop(promo)

df.reset_index(inplace=True, drop=True)
```

```python
import re
# =========================================================
# 문제2
#
# 확인해보니 레시피가 영어인 것들이 약 300개 정도 되는 것 같은데
# 제목이 모두 영어만 있는 것을 어떻게 찾을지 고민중
#
# 원래는 영어를 모두 공백으로 치환해서 공백인 행을 삭제하려고 함
# 문제 발생(LA갈비 => 갈비처럼 모든 영어가 다 삭제됨)
# =========================================================

def test(s):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
    result = hangul.sub('', s)
    return result
```

```python
# LA갈비는 고유명사처럼 쓰이는 데 
# 위의 test를 사용하면 '갈비'로만 나오는 문제가 생긴다.

test('LA갈비')
```
