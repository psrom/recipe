```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://www.10000recipe.com'
main_url = base_url + '/recipe/list.html?order=reco&page='
```

```python
resp = requests.get(main_url, params={
    'page': 1
})
```

```python
soup = BeautifulSoup(resp.text)
recipe_list = soup.select('a.common_sp_link')
recipe_list[0]
```

```python
url_list = []
for item in recipe_list:
    url_list.append(base_url+item.get('href'))
```

```python
url_list
```

```python
import tqdm
```

```python
content_list = []
```

```python
content_list = []
for sample in tqdm.tqdm(url_list):
    resp = requests.get(sample)
    content_list.append(resp.text)
```

## 저장하기

```python
import json
with open('./test.json','w') as f:
    json.dump(content_list, f)
```
