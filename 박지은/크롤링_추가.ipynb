# 크롤링

```python
image_url = []
tags = []
```

```python
for i in url_list[156650:156700]:
    res = requests.get(i)
    soup = BeautifulSoup(res.content, 'html.parser')

    tag_tag=[]
    
    if '10000recipe' in i:
      try:
        image = soup.find('div', 'centeredcrop').find('img')['src']
        image_url.append(image)
        tag = soup.find('div', 'view_tag').get_text()
        tag_tag.append(tag)
      except:
        tag_tag.append('#')

      tags.append(tag_tag)
      print(i)
      print(tag_tag)



    elif 'haemukja' in i:
      try:
        tag = soup.find('div', 'box_tag').get_text()
        #tag = tags.find_all('a').get_text()
        tag_tag.append(tag)
        #tags.append(tag_tag)
        if soup.find('div', 'flexslider') != None:

            image = soup.find('div', 'flexslider').find_all('img')[0]['src']
            image_url.append(image)
        else:    
            image = soup.find('ol','lst_step').find_all('div', 'img-cover')[-1].find('img')['src']
            image_url.append(image)    
      except:
        tag_tag.append('#')
      
      tags.append(tag_tag)
      print(i)
      print(tag_tag)


    else:
        x = url_list.index(i)
        tag = url_list_df['name'][x].split(" ")
        tags.append(tag)

        image = 'https://wtable.co.kr'+soup.find('div', style = "display:inline-block;max-width:100%;overflow:hidden;position:relative;box-sizing:border-box;margin:0").find_all('img')[1]['src']
        image_url.append(image)
        print(i)
        print(tag)
```

# 수집된 데이터 정리

```python
#길이가 모두 일치해야 한다
print(len(tags))
print(len(image_url))
```

```python
recipe_new_df = []
#[image_url, tags]
df = pd.DataFrame(recipe_new_df)
```

```python
df['image_url'] = image_url
df['tag'] = tags
```
