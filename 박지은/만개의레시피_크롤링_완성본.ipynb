# 1. 환경 설정

```python
import requests
from bs4 import BeautifulSoup
```

# 2. URL 정보

```python
#url = 'https://www.10000recipe.com/recipe/list.html?cat='
params = {
   'page' : 1
}



base_url = 'https://www.10000recipe.com'
main_url = base_url + '/recipe/list.html?order=reco&page='
```

# 3. 저장될 리스트 
## (페이지 범위 바뀌면 바뀔때마다 다시 실행)

```python
url_list = []
views_list=[]
```

```python
title = []
ingredients = []
peoples = []
times=[]
levels=[]
```

# 4. 크롤링 코드

```python
for i in range(1, 2):      #페이지 범위
    resp = requests.get(main_url, params={
            'page': i
            })
    print(resp)
    soup = BeautifulSoup(resp.content, 'html.parser')
    
    soup = BeautifulSoup(resp.text)
    recipe_list = soup.select('a.common_sp_link')
    vie = soup.select('span.common_sp_caption_buyer')
    for v in vie:
        view = v.get_text()
        print(view)
```

```python
for i in range(1, 2):      #페이지 범위
    resp = requests.get(main_url, params={
            'page': i
            })
    print(resp)
    soup = BeautifulSoup(resp.content, 'html.parser')
    
    soup = BeautifulSoup(resp.text)
    recipe_list = soup.select('a.common_sp_link')
    vie = soup.select('span.common_sp_caption_buyer')
    
    
    #url 수집
    for item in recipe_list:
        url_list.append(base_url+item.get('href'))
    
    #조회수 수집
    for v in vie:
        view = v.get_text()
        views_list.append(view)
        #print(view)
        
    for i in url_list:
        resp = requests.get(i)
        soup = BeautifulSoup(resp.content, 'html.parser')
    
    
        #제목
        title_lst = soup.find('h3').get_text()
        title.append(title_lst)
        print(title_lst)
    
    
        #재료
        ing_list = soup.find('div', class_ = "ready_ingre3")
        try:
            li_tags = ing_list.find_all('li')

            temp_ing_list =[]
            for ing_tag in li_tags:
                temp_ing_list.append(":".join(ing_tag.text.split()))
            ing = "|".join(temp_ing_list)
            ingredients.append(ing)
            print(ing)        
        except AttributeError:
            ingredients.append('@') #재료 없는것은 이렇게 처리
            print('재료없음')
        
        
        #인분, 조리시간, 난이도
        try:
            info_list = soup.find('div', class_ = 'view2_summary_info')
            #인분
            try:
                peop_list = soup.find('span', class_ = 'view2_summary_info1').get_text()
                peoples.append(','.join(peop_list.split()))
                print(peop_list)
            except AttributeError:
                peoples.append('@')
                print('인분정보없음')
          
            #조리시간 
            try:
                time_list = soup.find('span', class_ = 'view2_summary_info2').get_text()
                times.append(time_list)
                print(time_list)
            except AttributeError:
                times.append('@')
                print('조리시간정보없음')
            
            #난이도
            try:
                level_list = soup.find('span', class_ = 'view2_summary_info3').get_text()
                levels.append(level_list)
                print(level_list)
            except AttributeError:
                levles.append('@')
                print('난이도정보없음')
             
        except:
            #peoples.append('@')
            #times.append('@')
            levels.append('@')  #왜인지는 모르겠다...
            print('info없음')
```
